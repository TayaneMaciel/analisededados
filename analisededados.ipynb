#**Correlação**
***posiitiva*** - variaveis aumentam juntas.

***negativa*** - uma variavel aumenta, a outra diminui.


---
A função de correlação mais usada é a de **Pearson**, que varia de -1 a 1.
Um valor próximo de 1 indica uma forte correlação positiva, próximo de -1 indica uma forte correlação negativa, e próximo de 0 indica que não há uma relação linear.





```python
#Calculo da correlação de IDADE  e SALARIO [ usando função do PANDAS chamada corr]
correlacao_continua = dados['IDADE'].corr(dados['SALARIO'])
correlacao_continua
```

Para identificar correlações em tabelas com muitas colunas de valores contínuos, utilizamos **mapas de calor**.

Um exemplo de mapa de calor *permite visualizar rapidamente a correlação entre variáveis contínuas*, com cores indicando a força da correlação.

Para o cálculo de correlação entre variáveis categóricas, nós
podemos utilizar o coeficiente de Cramer, que basicamente é a normalização de 0 a 1 da
correlação das variáveis, onde zero indica nenhuma associação entre as variáveis e 1 indica uma
associação completa. Para calcular o coeficiente, vamos criar uma função que vai receber as
colunas que queremos a correlação.

```python
from scipy.stats import chi2_contingency
```

```python
def cramer_coeficiente(coluna1, coluna2):
  tabela_cruzada = np.array(pd.crosstab(coluna1,coluna2))
  chi2 = chi2_contingency(tabela_cruzada)[0]
  soma = np.sum(tabela_cruzada)
  mini = min(tabela_cruzada.shape)-1
  cramer = np.sqrt((chi2/(soma*(mini-1))))
  return cramer
```

```python
plt.figure()
plt.scatter(dados['IDADE'], dados['SALARIO'], alpha=0.5) #esse alpha define o grau de transparencia dos pontinhos
plt.xlabel('Idade')
plt.ylabel('Salario')
plt.title('Dispersao de salario por idade')
plt.grid(True)
plt.show()
```
